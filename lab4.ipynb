```python
import math
import cmath
import matplotlib.pyplot as plt
```

```python
# Передаточные функции

# Формула прямоугольниокв
def H_rectangle(w):
  return (
      cmath.exp(0.5j*w) /
      (cmath.exp(1j*w) - 1)
  )

# Формула трапеций
def H_trapezoid(w):
  return (
      (cmath.exp(1j*w) + 1) /
      (2 * (cmath.exp(1j*w) - 1))
  )

# Формула Симпсона
def H_simpson(w):
  return (
      (cmath.exp(1j*w) + 4 + cmath.exp(-1j*w)) /
      (3 * (cmath.exp(1j*w) - cmath.exp(-1j*w)))
  )

# 3/8
def H_3_8(w):
  return (
      (cmath.exp(2j*w) + 3*cmath.exp(1j*w) + 3 + cmath.exp(-1j*w)) /
      (8 * (cmath.exp(2j*w) - cmath.exp(-1j*w)))
  )
```

```python
# Отношение значений

# Формула прямоугольниокв
def Y_rectangle(w):
  return (
      (w / 2) / math.sin(w / 2)
  )

# Формула трапеций
def Y_trapezoid(w):
  return math.cos(w/2) * (
      (w / 2) / math.sin(w/2)
  )

# Формула Симпсона
def Y_simpson(w):
  return (
      ((2 + math.cos(w)) / 3) *
      (w / math.sin(w))
  )

# 3/8
def Y_3_8(w):
  return (
      (2 * math.cos(3*w/2) + 6 * math.cos(w/2)) * w /
      (16 * math.sin(3*w/2))
  )
```

```python
# Количество частот гармоник
N = 8192
# Циклические частоты
F = [f/N for f in range(1, N)]
# Круговые частоты
W = [2*math.pi * f for f in F ]

h_r = [H_rectangle(w) for w in W]
h_t = [H_trapezoid(w) for w in W]
h_s = [H_simpson(w) for w in W]
h_3_8 = [H_3_8(w) for w in W]

y_r = [Y_rectangle(w) for w in W]
y_t = [Y_trapezoid(w) for w in W]
y_s = [Y_simpson(w) for w in W]
y_3_8 = [Y_3_8(w) for w in W]
```

```python
plt.plot(F, [x.imag for x in h_r], label='$Im(H_{rectangle})$', color='royalblue')
plt.plot(F, [x.imag for x in h_t], label='$Im(H_{trapezoid})$', color='crimson')
plt.plot(F, [x.imag for x in h_s], label='$Im(H_{simpson})$', color='gold')

plt.xlabel('f')
plt.ylabel('$\widetilde{H}(f)$')
plt.legend()
plt.ylim((-3, 3))
plt.grid()
plt.show()
!
```

```python
plt.plot(F, y_r, label='$\gamma_{rectangle}$', color='royalblue')
plt.plot(F, y_t, label='$\gamma_{trapezoid}$', color='crimson')
plt.plot(F, y_s, label='$\gamma_{simpson}$', color='gold')

plt.xlabel('f')
plt.ylabel('$\gamma$')
plt.legend()
plt.ylim((-3, 3))
plt.grid()
plt.show()
!
```
